```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(32)

u = 5
sigma = 1

array = np.random.randn(1000)
vals = u + sigma * array
```

```python
plt.figure(figsize=(8, 6))

plt.legend("Plot Histogram")
plt.xlabel('Bins')
plt.ylabel("Frequency")

plt.hist(vals)

plt.show()
```

```python
z = np.unique(np.round(vals), return_counts=True)
```
