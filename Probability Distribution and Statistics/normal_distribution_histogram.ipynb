```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(32)

u = 5
sigma = 2

array = np.random.randn(10000)
array = u + sigma * array
```

```python
plt.figure(figsize=(8, 6))

plt.legend("Plot Histogram")
plt.xlabel('Bins')
plt.ylabel("Frequency")

plt.hist(array)

plt.show()
```
