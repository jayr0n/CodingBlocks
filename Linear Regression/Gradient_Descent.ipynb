```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2

print(X, Y)
```

The **goal** is to find the value of x that minimize f(x)

```python
plt.plot(X, Y)
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

```python
x = 0
lr = 0.1
error = []

plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x, y)
    #print(x)
    
```

```python
plt.plot(error)
plt.show()
```
