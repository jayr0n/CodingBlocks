```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
image1 = cv2.imread("C:/Users/Jayron/Documents/coding blocks/bottom_left.jpg")
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image2 = cv2.imread("C:/Users/Jayron/Documents/coding blocks/bottom_right.jpg")
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

image3 = cv2.imread("C:/Users/Jayron/Documents/coding blocks/center.jpeg")
image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)

image4 = cv2.imread("C:/Users/Jayron/Documents/coding blocks/top_left.jpg")
image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

image5 = cv2.imread("C:/Users/Jayron/Documents/coding blocks/top_right.jpg")
image5 = cv2.cvtColor(image5, cv2.COLOR_BGR2RGB)

image1_resized = cv2.resize(image1, (200, 200))
image2_resized = cv2.resize(image2, (200, 200))
image3_resized = cv2.resize(image3, (200, 200))
image4_resized = cv2.resize(image4, (200, 200))
image5_resized = cv2.resize(image5, (200, 200))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def centralizar_imagem(base_image, central_image):
    base_h, base_w, _ = base_image.shape
    center_h, center_w, _ = central_image.shape
    
    start_h = (base_h - center_h) // 2
    start_w = (base_w - center_w) // 2
    
    resultado = base_image.copy()
    
    resultado[start_h:start_h + center_h, start_w:start_w + center_w] = central_image
    
    return resultado

border_size = 10
image1_with_border = np.pad(image1_resized, ((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=0)
image2_with_border = np.pad(image2_resized, ((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=0)
image3_with_border = np.pad(image3_resized, ((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=0)  # IMAGEM DO MEIO
image4_with_border = np.pad(image4_resized, ((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=0)
image5_with_border = np.pad(image5_resized, ((border_size, border_size), (border_size, border_size), (0, 0)), mode='constant', constant_values=0)

top_row = np.concatenate((image4_with_border, image5_with_border), axis=1)
bottom_row = np.concatenate((image1_with_border, image2_with_border), axis=1)

final_image = np.concatenate((top_row, bottom_row), axis=0)

image3_resized = cv2.resize(image3_resized, (150,150))
imagem_final_com_centro = centralizar_imagem(final_image, image3_resized)

plt.imshow(imagem_final_com_centro)
plt.axis('off')
plt.show()
```

```python
pixels = imagem_final_com_centro.reshape(-1, 3)

# Converter para DataFrame com os nomes das colunas 'r', 'g', 'b'
df = pd.DataFrame(pixels, columns=['r', 'g', 'b'])

#df.to_csv('C:/Users/Jayron/Documents/coding blocks/submission.csv', index=False)
```
